```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/referrer-monthly-top100_2.csv')
labels = df.iloc[2:12,:]
labels['count.1']
# labels.head()
```

```python
indexes = {}
arrs = []

arlen = 0
for host in range(0,106):
    num = ''
    if host != 0:
        num = '.'+str(host)
    
    hostCol = 'host'+num
    countCol = 'count'+num

    for i, url in enumerate(labels[hostCol]):
        if url not in indexes:
            if len(arrs) == 0:
                indexes[url] = 0
                arrs.append([])
            else:
                arrs.append([nan for j in range(0,arlen-1)])
                indexes[url] = len(arrs)-1
            
        arrs[indexes[url]].append(labels[countCol][i+2])
        arlen = len(arrs[indexes[url]])
    
    for val in arrs:
        if len(val) < arlen:
            val.append(0)
```
