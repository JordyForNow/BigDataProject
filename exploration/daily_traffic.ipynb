```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv("../Data/count_per_day_final.csv")
data.head()
```

```python
dates = data.iloc[:,0]
counts = data.iloc[:,1]
xTicks = []
tickLocs = []
prev = "3"

for i, date in enumerate(dates):
    if date[-1] != prev:
        prev = date[-1]
        xTicks.append(date)
        tickLocs.append(i)

xTicks.append(dates.iloc[-1])
tickLocs.append(len(dates)-1)
```

```python

figure(figsize=(10, 4), dpi=80, layout='tight')
plt.plot(counts)
plt.xticks(ticks=tickLocs, labels=xTicks, rotation=90)
```

```python
hist = counts.to_list()
hist.sort()
```

```python
figure(figsize=(6, 4), dpi=80, layout='tight')
plt.hist(hist, bins=100)
```
